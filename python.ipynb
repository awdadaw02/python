```python
import random
import time
import matplotlib as plt
def f1(n):
    x1=[]
    for x in range(n):
        x1=x1+[random.random()]
    return x1    
        

def f2(n):
    x1=[random.random() for x in range(n)]
    return x1
```

```python
nn=10000
t1=time.time()
f1(nn)
t2=time.time()
print(t2-t1)

```

```python
t1=time.time()
f2(nn)
t2=time.time()
print(t2-t1)
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from multiprocessing import Pool
import time


nn=10000
y1=[]
y2=[]
x=[x for x in range(100,10000, 100)]

for n  in  x:
    t1 =time.time()
    f1(nn)

    t2 =time.time()
    y1.append(t2-t1)
    t1 =time.time()
    f2(nn)
    t2 =time.time()
    y2.append(t2-t1)

%matplotlib inline

plt.plot(x, y1,  color='#FB8500', label='Multi')
plt.plot(x, y2,  color='#023047', label='Single')
plt.legend()
plt.title('Оценка производительности: чем выше - тем хуже')
plt.show()
```
